```python
from graphviz import Digraph

# Create a new directed graph
dot = Digraph()

# Define the input, process, and output nodes
dot.node('A', 'User Input', shape='oval')  # Input layer
dot.node('B', 'Data Preprocessing', shape='box')  # Process layer
dot.node('C', 'Load Model', shape='box')  # Process layer
dot.node('D', 'Predict Churn', shape='box')  # Process layer
dot.node('E', 'Output Prediction', shape='oval')  # Output layer

# Add edges between the nodes
dot.edge('A', 'B', label='Customer Details\n(tenure, monthly charges,\ncontract type, payment method)')
dot.edge('B', 'C', label='Preprocessed Data')
dot.edge('C', 'D', label='Trained Model')
dot.edge('D', 'E', label='Churn Prediction\n(Yes/No)')

# Render the graph to a file (you can also view it in a Jupyter Notebook)
dot.render('customer_churn_dfd', format='png', cleanup=True)  # Output as PNG
dot.render('customer_churn_dfd', format='pdf', cleanup=True)  # Output as PDF

# Optionally, print the source code for reference
print(dot.source)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Generate sample resume data
np.random.seed(42)  # Ensures consistent results
resumes = [f"Resume {i}" for i in range(1, 11)]  # 10 sample resumes
scores = np.random.randint(40, 95, size=10)  # Random scores between 40% - 95%

# Create a DataFrame
results = pd.DataFrame({"Resume": resumes, "Score": scores})

# Sort results for better visualization
sorted_results = results.sort_values(by="Score", ascending=True)

# Create a figure
fig, ax = plt.subplots(figsize=(8, 5))
sns.barplot(y=sorted_results["Resume"], x=sorted_results["Score"], palette="Blues_r", ax=ax)

ax.set_xlabel("Score (%)")
ax.set_ylabel("Resumes")
ax.set_title("Resume Ranking Scores")
ax.set_xlim(0, 100)

# Show values on bars
for i in ax.containers:
    ax.bar_label(i, fmt='%.1f%%', label_type="edge", fontsize=10)

# Show the plot
plt.show()
```

```python
import pandas as pd
import folium
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("cleaned_dataset.csv")

# Drop rows with NaN values in 'Latitude' or 'Longitude' columns
df = df.dropna(subset=['Latitude', 'Longitude'])

m = folium.Map(location=[df['Latitude'].mean(), df['Longitude'].mean()], zoom_start=10)

for index, row in df.iterrows():
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=3,
        color='red',
        fill=True,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(m)

m.save("restaurant_locations_map.html")
print("Map saved as restaurant_locations_map.html")
```

```python
import matplotlib.pyplot as plt

def plot_model_performance(emotion_score, speech_score, final_score):
    # Bar chart for performance visualization
    categories = ['Emotion Score', 'Speech Score', 'Final Confidence Score']
    scores = [emotion_score, speech_score, final_score]

    plt.bar(categories, scores, color=['blue', 'orange', 'green'])
    plt.title("Model Performance")
    plt.xlabel("Score Categories")
    plt.ylabel("Scores")
    plt.ylim(0, 100)  # Assuming score range between 0 and 100
    plt.show()

# Example usage
plot_model_performance(80, 70, 75)
```
